```python
import pandas as pd
import json 
```

```python
with open ("t20_json_files/t20_wc_match_results.json") as d:
    data = json.load(d)
df_match = pd.DataFrame(data[0]["matchSummary"])
```

## 1) Processing match summary

```python
df_match.head()
```

```python
df_match.shape
```

## Renaming scorecard as match_id 

```python
df_match.rename({"scorecard" : "match_id"}, axis = 1 ,inplace = True)
```

```python
df_match.head()
```

## checking null values


```python
df_match.isnull().sum()
```

## Create a dictonary that maps VsTeam name to a unique match id 

```python
match_id_dict = {}
for index, row in df_match.iterrows():
    key_1 = row["team1"] + " Vs " + row["team2"]
    key_2 = row["team2"] + " Vs " + row["team1"]
    match_id_dict[key_1] = row["match_id"]
    match_id_dict[key_2] = row["match_id"]
```

```python
match_id_dict
```

## Exporting processed file to csv file


```python
df_match.to_csv("t20_csv_files/dim_match_summary.csv", index = False)
```

## 2) Processing batting summary

```python
with open ("t20_json_files/t20_wc_batting_summary.json") as d:
    data = json.load(d)
    
all_records = []

for rec in data:
    all_records.extend(rec["battingSummary"])
```

## converting to panda dataframe for further processing


```python
df_batting = pd.DataFrame(all_records)
```

```python
df_batting.head()
```

```python
df_batting.shape
```

## Adding column if player out or not out 

```python
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
```

```python
df_batting.head(20)
```

## removing column how players got out

```python
df_batting.drop("dismissal" , axis = 1,  inplace = True)
```

```python
df_batting
```

```python
df_batting.head(10)
```

## removing special characters in column

```python
df_batting['batsmanName'] = df_batting['batsmanName'].str.replace(r"[^a-zA-Z0-9\s\[\]\(\)\{\}-]", "", regex=True)
```

```python
df_batting.head(10)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_id_dict)
```

```python
df_batting.isnull().sum()
```

```python
df_batting.head()
```

## Exporting processed file to csv file


```python
df_batting.to_csv("t20_csv_files/fact_bating_summary.csv", index = False)
```
